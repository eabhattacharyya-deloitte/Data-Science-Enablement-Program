<b>Basic Plot Creation with Matplotlib</b>


```python

import matplotlib.pyplot as plt
import numpy as np

# Dataset for examples
x = np.arange(1, 6)
y = np.array([5, 8, 12, 7, 10])


```

```python
# 1. Bar Chart
plt.figure(figsize=(6, 4))
plt.bar(x, y)
plt.title('Simple Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

```

```python
# 2. Line Chart
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o')
plt.title('Simple Line Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

```

```python
# 3. Scatter Plot
np.random.seed(0)
x_scatter = np.random.rand(50)
y_scatter = 2 * x_scatter + np.random.randn(50) * 0.2

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, color='green')
plt.title('Simple Scatter Plot')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.show()
```

<b>Customizing Charts: Titles, Axis Labels, Legends, and Styles</b>


```python

x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(7, 5))
plt.plot(x, y_sin, label='Sine', color='navy', linewidth=2, marker='o')
plt.plot(x, y_cos, label='Cosine', color='orange', linestyle='--', linewidth=2)
plt.title('Sine and Cosine Waves', fontsize=16)
plt.xlabel('Angle [radians]', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, linestyle=':', color='gray', alpha=0.6)
plt.tight_layout()
plt.show()
```

<b>Visualization Best Practices: Clarity, Chart Choice, and Honest Storytelling</b>

```python
import pandas as pd

# Example data: Monthly sales (in thousands) by two regions
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
north = [90, 110, 115, 130, 125]
south = [120, 105, 98, 140, 135]

# Side-by-side Bar Chart for Clear Regional Comparison
x = np.arange(len(months))
width = 0.35

plt.figure(figsize=(8, 4))
plt.bar(x - width/2, north, width=width, label='North', color='#007CB0')
plt.bar(x + width/2, south, width=width, label='South', color='#86BC25')
plt.title('Monthly Sales by Region')
plt.xlabel('Month')
plt.ylabel('Sales (thousands)')
plt.xticks(x, months)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Notes on Best Practice:
# - Compare data fairly (side-by-side bars prevent misinterpretation)
# - Always use clearly labeled axes and legends
# - Avoid 3D effects, excessive colors, or chartjunk
# - Choose chart types that fit your message: 
#   (e.g., use line for trends, bar for categories, scatter for relationships)
```
