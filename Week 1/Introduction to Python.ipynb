<b>ü§î Chapter 5: Making Decisions</b>

```python
# Simple decision making
age = int(input("How old are you? "))

if age >= 18:
    print("You're an adult!")
else:
    print("You're still young!")
```

```python
# Multiple choices
temperature = int(input("What's the temperature outside? "))

if temperature > 80:
    print("It's hot! Wear shorts.")
elif temperature > 60:
    print("Nice weather! Perfect for a walk.")
elif temperature > 40:
    print("A bit chilly. Bring a jacket.")
else:
    print("It's cold! Bundle up!")
```

```python
# Combining conditions
age = 20
has_license = True

if age >= 16 and has_license:
    print("You can drive!")
elif age >= 16 and not has_license:
    print("You're old enough, but you need a license.")
else:
    print("You're too young to drive.")
```

<b>üîÑ Chapter 6: Repeating Actions (Loops)</b>

```python
# Count from 0 to 4
print("Counting:")
for i in range(5):
    print("Count:", i)
```

```python
print("\nCounting from 1 to 10:")
for i in range(1, 11):
    print(i)
```

```python
# Count by 2s
print("\nEven numbers:")
for i in range(2, 11, 2):
    print(i)
```

```python
# Working with lists of things
fruits = ["apple", "banana", "orange", "grape"]
print("\nI like these fruits:")
for fruit in fruits:
    print("üçé", fruit)
```

```python
# Counting with items
print("\nNumbered list:")
for i, fruit in enumerate(fruits, 1):
    print(f"{i}. {fruit}")
```

```python
# Keep asking until they get it right
secret_word = "python"
guess = ""

print("Guess the secret word!Hint: It is a snake")
while guess != secret_word:
    guess = input("Your guess: ").lower()
    if guess != secret_word:
        print("Nope, try again!")

print("üéâ You got it!")
```

```python
# Countdown
print("\nCountdown:")
count = 5
while count > 0:
    print(count)
    count = count - 1
print("üöÄ Blast off!")
```

<b>üõ†Ô∏è Chapter 7: Creating Your Own Tools (Functions)</b>


Functions are like recipes - you write them once and use them many times!

```python
# Simple function
def say_hello():
    print("Hello there!")
    print("How are you today?")

# Use the function
say_hello()
```

```python

# Function that takes information (parameters)
def greet_person(name):
    print(f"Hello, {name}!")
    print("Welcome to Python!")

greet_person("Alice")
greet_person("Bob")
```

```python
# Function that gives back a result
def add_numbers(first, second):
    result = first + second
    return result

# Use the function
sum1 = add_numbers(5, 3)
sum2 = add_numbers(10, 20)
print("5 + 3 =", sum1)
print("10 + 20 =", sum2)
```

üéØ Practical Functions

```python
# Calculate area of a rectangle
def calculate_area(length, width):
    area = length * width
    return area

# Convert temperature
def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius

# Check if a number is even
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False

# Test our functions
room_area = calculate_area(12, 10)
print(f"Room area: {room_area} square feet")

temp_c = fahrenheit_to_celsius(75)
print(f"75¬∞F is {temp_c:.1f}¬∞C")

print("Is 8 even?", is_even(8))
print("Is 7 even?", is_even(7))
```
